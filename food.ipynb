```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import re
import time

BASE_URL = "https://www.fatsecret.kr"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36",
}

# ✅ 음식 리스트 (requests)
def get_food_list(page=1):
    url = f"{BASE_URL}/칼로리-영양소/?page={page}"
    resp = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")
    foods = soup.select("table.generic.common a.prominent")
    data = []
    for f in foods:
        name = f.get_text(strip=True)
        link = BASE_URL + f["href"]
        data.append({"음식명": name, "링크": link})
    return data

# ✅ Selenium으로 상세 정보 (브라우저 1개만 사용)
def get_food_details(driver, food_links):
    results = []

    for food in food_links:
        driver.get(food["링크"])
        time.sleep(1.2)  # 로딩 대기

        rows = driver.find_elements("css selector", "table tr")
        cal, protein = None, None
        for row in rows:
            cells = row.find_elements("tag name", "td")
            if len(cells) >= 2:
                label = cells[0].text.strip()
                value = cells[1].text.strip()
                if "칼로리" in label:
                    cal = re.sub(r"[^0-9]", "", value)
                elif "단백질" in label:
                    protein = re.sub(r"[^0-9]", "", value)

        results.append({
            "음식명": food["음식명"],
            "칼로리": int(cal) if cal else None,
            "단백질": int(protein) if protein else None,
            "링크": food["링크"]
        })
    return results
```

```python

# ✅ 메인 실행
def main():
    # 음식 목록 수집
    food_links = []
    for page in range(1, 3):  # 테스트용 2페이지
        food_links.extend(get_food_list(page))
    print(f"총 {len(food_links)}개의 음식 수집 완료.")

    # 브라우저 1번만 실행
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    # 상세 정보 크롤링
    results = get_food_details(driver, food_links)
    driver.quit()

    # CSV 저장
    df = pd.DataFrame(results)
    df.to_csv("fatsecret_fast.csv", index=False, encoding="utf-8-sig")
    print("✅ 전체 데이터 저장 완료.")

    # 단백질 ≥ 30g, 칼로리 ≤ 500 kcal 필터링
    filtered = df[(df["단백질"] >= 30) & (df["칼로리"] <= 500)]
    filtered.to_csv("fatsecret_filtered.csv", index=False, encoding="utf-8-sig")
    print("✅ 필터링된 데이터 저장 완료.")
```
